```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mercdtf=pd.read_excel("merc.xlsx")
sıralanmısmerc=mercdtf.sort_values("price",ascending=False).iloc[131:]
sıralanmısmerc=sıralanmısmerc[sıralanmısmerc.year!=1970]
sıralanmısmerc=sıralanmısmerc.drop("transmission",axis=1)
```

```python
y=sıralanmısmerc["price"].values
```

```python
x=sıralanmısmerc[["year","mileage","mpg","engineSize"]].values
```

```python
x1,x2,y1,y2=train_test_split(x,y,test_size=0.33,random_state=10)
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x1)
```

```python
x1=scaler.transform(x1)
```

```python
x2=scaler.transform(x2)
```

```python
model=Sequential()
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x1,y=y1,validation_data=(x2,y2),batch_size=250,epochs=500)
```

```python
loss=pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
dizi=model.predict(x2)
```

```python
mean_absolute_error(y2,dizi)
```
